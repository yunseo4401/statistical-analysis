#교차분석 
- 카이제곱검정은 설명변수가 범주형 변수인 경우에 사용한다
- 데이터에서 각 범주에 따른 결과변수의 분포를 설명하거나, 범주형 변수가 두개 이상인 경우 두 변수가 상관이 있는지를 검정하는 것이 목적이다. 
- 카이제곱 검정은 검정의 목적에 따라 사용되는 3가지 검정이 있음 : 적합성검정, 독립성 검정, 동질성 검정 

#1. 적합성 검정 
- 각 범주에 따른 데이터의 빈도분포가 이론적으로 기대하는 분포를 따르는지 검정 
ex) 주사위를 굴렀을 때, 각 주사위의 값이 1/6이 맞는지를 검정 


```
scipy.stats.chisquare(f_obs,f_exp=None,ddof=0,axis=0)
```
-f_obs : 관찰빈도로, pd.value_counts() 결과 값을 입력

-f_exp : 각 카테고리의 기대빈도 

-ddof : p-value에 대한 자유도를 조정가능 



````python
import pandas as pd
df=pd.read_csv('titanic.txt',sep=',')
df.info()
````

````python
df
#sex변수에 대한 분할표를 생성
#귀무가설 : 생존자 비율중 남자의 비율이 50, 여자의 비율이 50이다. 
#대립가설 : ~~라고 할 수없다. 
````

````python
df_t=df[df['survived']==1]
table=df_t[['sex']].value_counts()
table
````

````python
#sex변수에 대한 도수분포표를 생성하여 범주별 관찰빈도를 확인
from scipy.stats import chisquare 
chi=chisquare(table,f_exp=[171,171])
print('적합도 검정',chi)
#귀무가설 기각 > 비율 50:50아님 
````

#2. 독립성 검정 
- 모집단이 두 개의 변수 A,B에 의해 범주화되었을때, 이 두변수들 사이의 관계가 독립인지 아닌지 검정


```
scipy.stats.chi2_contigency((observed,correction=True,lambda_=None)))
```
-observed : 관찰빈도로 pd.crosstab 결과 값을 입력 


````python
#좌석등급(class)와 생존 여부가 서로 독립인지 검정해보자 
#귀무가설 : class와 survived는 독립이다 
#대립가설 : 독립 아님 
table=pd.crosstab(df['class'],df['survived'])
table
````

````python
from scipy.stats import chi2_contingency
chi,p,df,expect = chi2_contingency(table) #카이제곱 통계량, p-value,자유도, 테이블 합계를 기반으로 힌 기대빈도 
print(chi,p,df,expect)
#귀무가설 기각, 독립아님 
````
